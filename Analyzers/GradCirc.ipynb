```python
import numpy as np
import torch as t
import matplotlib.pyplot as plt
import matplotlib
import numpy.polynomial.polynomial as poly
%matplotlib inline
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
data = np.load('/Users/jonathanlu/Documents/data_sym/Reg.npy')
x_raw = data[0]
y_raw = 1/data[1]
x = np.linspace(0, max(x_raw), 101)
print(max(x_raw))
coeffs = poly.polyfit(x_raw, y_raw, 5)
fit = poly.polyval(x, coeffs)
```

```python
FONT_SIZE = 17
plt.scatter(x_raw, y_raw, c='silver', alpha=0.75, s=15)
plt.plot(x, fit, color='navy')
plt.xlim(0, 5)
plt.xlabel("Step size", fontsize=FONT_SIZE)
plt.ylabel(r"Local learning time", fontsize=FONT_SIZE)
plt.xticks(fontsize=FONT_SIZE)
plt.yticks(fontsize=FONT_SIZE)
plt.tight_layout()
plt.savefig('/Users/jonathanlu/Documents/data_sym/Regularizer.pdf')
```
